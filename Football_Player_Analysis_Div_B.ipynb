# Borrussia Dortmund is seeking a valuable striker under the age of 22 for their team. Your responsibility is to create a shortlist of players for them to consider who are under the value of 40 Million Euros. 


## 1st Step: Brainstorm Which Columns/Criteria You Will Need to Consider Players

## 2nd Step: Finalize Which Columns/Criteria You Will Use

## 1. Making Our Shortlist Only With Our Selected Columns

```python
# Keeping only the columns we need from players_df in our shortlist_df
short_list_df = players_df[["sofifa_id",
                            'long_name',
                            "age",
                            "league_name",
                            "team_position", 
                            'preferred_foot',
                            'player_traits',
                            'wage_eur',
                            "value_eur",
                            'contract_valid_until',
                            'release_clause_eur',
                            "power_stamina",
                            "pace",
                            "shooting", 
                            "skill_dribbling",
                            'skill_curve',
                            "attacking_finishing",
                            'attacking_heading_accuracy',
                            "overall",
                            'potential',
                            'player_url']]

# Printing the first 5 rows of shortlist_df
short_list_df.head()
```

## 2. Adding our Minimum Criteria

- Age: 22
- Value: 40 Million Euros
- Contract Expiry: 2023
- Release Clause <= 40 Million Euros 
- Position: LS, ST, RS, LF, RF, CF, 

```python
# Filtering our data based on the initial minimum criteria
short_list_df = short_list_df.loc[(short_list_df["age"]<=22) 
                                 & (short_list_df["value_eur"]<=40000000)
                                 & (short_list_df["contract_valid_until"] <=2023)
                                 & (short_list_df["release_clause_eur"]<=40000000)
                                 & ((short_list_df["team_position"] == "ST") | 
                                   (short_list_df["team_position"] == "LS") | 
                                   (short_list_df["team_position"] == "RS") | 
                                   (short_list_df["team_position"] == "LF") |
                                   (short_list_df["team_position"] == "CF") | 
                                   (short_list_df["team_position"] == "RF")  )]

short_list_df
```

```python
# Identified the median values for the player metrics
short_list_df.describe()
```

## 3. Adding Further Minimum Criteria
- Pace: 71
- Shooting: 63
- Dribbling: 64
- Finishing: 65

```python
# Further Filtering our data based on the minimum metric criteria
short_list_df = short_list_df.loc[(short_list_df["pace"]>=71)
                                 & (short_list_df["shooting"]>=63)
                                 & (short_list_df["skill_dribbling"]>=64)
                                 & (short_list_df["attacking_finishing"]>=65)]
short_list_df
```

## 4. Adding Columns Describing Special Traits
- Skill Curve: 70
- Heading Accuracy:70

```python
# Adding Columns About Special Traits of the Player
short_list_df.loc[(short_list_df["skill_curve"]>=70), "flair"] = "Yes"
short_list_df.loc[(short_list_df["skill_curve"]<70), "flair"] = "No"
short_list_df.loc[(short_list_df["attacking_heading_accuracy"]>=70), "good_header"] = "Yes"
short_list_df.loc[(short_list_df["attacking_heading_accuracy"]<70), "good_header"] = "No"

# Printing our final shortlist in descending order of potential, overall, and value
short_list_df.sort_values(["potential", "overall", "value_eur"], ascending = False)
```

```python
# Saving our shortlist to be sent to the Technical Director
short_list_df.to_csv("Striker_Shortlist.csv")
```

# Formula

## Sid Index: 
Average of this (Dribbling/Finishing/Pace/Stamina)  ... we can also consider potential after ... we measure how far they are from the existing 75th percentile value

## Arsh Index: 
Average of (Pace/Stamina/Potential Growth/Agility)

```python
short_list_df = players_df[["sofifa_id",
                            'long_name',
                            "age",
                            "league_name",
                            "team_position", 
                            'preferred_foot',
                            'player_traits',
                            'wage_eur',
                            "value_eur",
                            'contract_valid_until',
                            'release_clause_eur',
                            "power_stamina",
                            "pace",
                            "shooting", 
                            "skill_dribbling",
                            'skill_curve',
                            "attacking_finishing",
                            'attacking_heading_accuracy',
                            "overall",
                            'potential',
                            'player_url']]


short_list_df = short_list_df.loc[(short_list_df["age"]<=22) 
                                 & (short_list_df["value_eur"]<=40000000)
                                 & (short_list_df["contract_valid_until"] <=2023)
                                 & (short_list_df["release_clause_eur"]<=40000000)
                                 & ((short_list_df["team_position"] == "ST") | 
                                   (short_list_df["team_position"] == "LS") | 
                                   (short_list_df["team_position"] == "RS") | 
                                   (short_list_df["team_position"] == "LF") |
                                   (short_list_df["team_position"] == "CF") | 
                                   (short_list_df["team_position"] == "RF")  )]

short_list_df
```

# Approach to making  a formula

```python
short_list_df["z_index"]= (short_list_df["skill_dribbling"]+short_list_df["attacking_finishing"])
short_list_df["y_index"]= (short_list_df["pace"]* 0.5) * (short_list_df["skill_dribbling"]+short_list_df["attacking_finishing"])
short_list_df["x_index"]= (short_list_df["pace"]* 0.5) * (short_list_df["skill_dribbling"]+short_list_df["attacking_finishing"]+(short_list_df["shooting"]* 0.6)) * (short_list_df["power_stamina"] * 0.2)


short_list_df["growth_percentage"]=((short_list_df["potential"]- short_list_df["overall"])/short_list_df["overall"]) * 100
short_list_df["x_index_future"]= short_list_df["x_index"]*(100 + short_list_df["growth_percentage"])


short_list_df.sort_values(["x_index"] , ascending= False)
```

```python
# We want to normalize our data to a scale of 0 to 100 which is easy to read
# formula to normalise the data is
# ((value-min)/(max-min)) * 100


# normalising the value to be between 0 to 100
short_list_df["x_index_future_normalized"]=((short_list_df["x_index_future"]-
                                            short_list_df["x_index_future"].min())/
                                           (short_list_df["x_index_future"].max()-
                                           short_list_df["x_index_future"].min()))*100

short_list_df.sort_values(["x_index_future_normalized"], ascending= False).head(10)
```
